```python
import numpy as np

embed = {}
with open("../input/glove6b100dtxt/glove.6B.100d.txt",'r', encoding = 'utf-8') as f:
    for txt in f:
        a = txt.split()
        key, value = a[0], a[1:]
        embed[key] = np.asarray(value)

def func_c(x,y):
    a = np.squeeze(np.asarray(embed[x], dtype=float).reshape(1,-1))
    b = np.squeeze(np.asarray(embed[y], dtype=float).reshape(1,-1))
    #print(a)
    print(x + ' & '+y +" : "+str(cosine_sim(a, b)) + '\n')
    
 #cosine similarities
def cosine_sim(a,b):
    numerator = np.dot(a,b,out=None)
    denom= np.sqrt(sum(np.square(a)))*np.sqrt(sum(np.square(b)))
    result = round(numerator/denom, 4)
    return result


#man &woman
func_c('man', 'woman')

#cat & dog
func_c('cat', 'dog')

#king & queen
func_c('king', 'queen')



#king-man+woman
king = np.squeeze(np.asarray(embed['king'], dtype=float).reshape(1,-1))
man = np.squeeze(np.asarray(embed['man'], dtype=float).reshape(1,-1))
woman = np.squeeze(np.asarray(embed['woman'], dtype=float).reshape(1,-1))

#queen
queen = np.squeeze(np.asarray(embed['queen'], dtype=float).reshape(1,-1))

res = king-man+woman


print('queen & res : ' +str(cosine_sim(queen, res)) + '\n')

```
